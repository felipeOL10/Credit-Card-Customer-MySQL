```python
import matplotlib.pyplot as plt
import seaborn as sns
print("Setup Complete")
```

```python
# Most lucrative sectors
plt.figure(figsize=(30,10))
plt.title("EBITDA per sector")
sns.barplot(x=data.Sector, y=data['EBITDA'])
plt.ylabel("EBITDA")
plt.show()
```

```python
#  P/E per sector
plt.figure(figsize=(30,10))
plt.title("Price/Earnings per sector")
sns.barplot(x=data.Sector, y=data['Price/Earnings'])
plt.ylabel("P/E")
plt.show()
```

```python
#  P/S per sector
plt.figure(figsize=(30,10))
plt.title("P/S per sector")
sns.barplot(x=data.Sector, y=data['Price/Sales'])
plt.ylabel("P/S")
plt.show()
```

```python
#  Dividend yield per sector
plt.figure(figsize=(30,10))
plt.title("Dividend Yield per sector")
sns.barplot(x=data.Sector, y=data['Dividend Yield'])
plt.ylabel("Dividend Yield")
plt.show()
```

```python
sns.kdeplot(data=data['Price/Earnings'], label="Price/Earnings", shade=True)
```

```python
from sklearn import linear_model
import statsmodels.api as sm
```

```python
X = data[["Dividend Yield","Price/Sales", "Earnings/Share"]]
y = data["Price"]
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
predictions = model.predict(X) 
print_model = model.summary()
print(print_model)
```
